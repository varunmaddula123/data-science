```python
df['vehicletype'].fillna(df['vehicletype'].mode()[0],inplace=True)
df['gearbox'].fillna(df['gearbox'].mode()[0],inplace=True)
df['model'].fillna(df['model'].mode()[0],inplace=True)
df['fueltype'].fillna(df['fueltype'].mode()[0],inplace=True)
df['notrepaireddamage'].fillna(df['notrepaireddamage'].mode()[0],inplace=True)
```

```python
df.isnull().sum()
```

```python
df['price']
```

```python
import re
def prepor(price):

  price= re.sub(r"[^0-9]", "", price)
  return price
```

```python
df['newprice']=df['price'].apply(prepor)
```

```python
df
```

```python
df['newprice']=pd.to_numeric(df['newprice'])
df['datecrawled']=pd.to_datetime(df['datecrawled'])
df['datecreated']=pd.to_datetime(df['datecreated'])
```

```python
X = df.select_dtypes(include=['number'])
y = df['newprice']
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test =train_test_split(X,y,test_size=0.2,random_state=98)
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
model =LinearRegression()
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)



print(f"Mean Squared Error: {mse}")
print(f"R^2 Score: {r2}")
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
plt.hist(df['newprice'], bins=50, range=(0, 50000))
plt.title('Distribution of Car Prices')
plt.xlabel('Price (€)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Registration year analysis
plt.figure(figsize=(10,6))
plt.hist(df['yearofregistration'], bins=500)
plt.title('Distribution of Registration Years')
plt.xlabel('Year')
plt.ylabel('Frequency')
plt.show()
```

```python
# Brand analysis
brand_counts = df['brand'].value_counts().head(10)
brand_counts.plot(kind='bar', figsize=(10,6))
plt.title('Top 10 Car Brands by Listings')
plt.ylabel('Number of Listings')
plt.show()

# Average price by brand
avg_price = df.groupby('brand')['newprice'].mean().sort_values(ascending=False).head(10)
avg_price.plot(kind='bar', figsize=(10,6))
plt.title('Top 10 Brands by Average Price')
plt.ylabel('Average Price (€)')
plt.show()
```
